```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from tensorflow import keras
```

```python
train_ds = keras.utils.image_dataset_from_directory(
    directory = 'C:/Users/Deepak Santhosh/Documents/Deep_learning/Plants',
    label_mode = 'int',
    batch_size=32,
    image_size=(256,256)
)

```

```python
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label

train_ds = train_ds.map(process)
```

```python
from keras.layers import Dense,Flatten,MaxPooling2D,Conv2D,Dropout,BatchNormalization
from tensorflow import keras
from keras.models import Sequential
```

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(256,256,3)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(64,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Conv2D(128,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(4,activation='softmax'))
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

```python
model.compile(loss=loss,metrics=['accuracy'],optimizer='adam')
```

```python
model.fit(train_ds,epochs=5)
```

```python
def predict(image_path):
  img = cv2.imread(image_path)
  resize = tf.image.resize(img,(256,256))
  yhat = model.predict(np.expand_dims(resize/255,0))
  return np.argmax(yhat),yhat
```

```python
from pathlib import Path
import imghdr

data_dir = "C:/Users/Deepak Santhosh/Documents/Deep_learning/Plants"
image_extensions = [".png", ".jpg"]  # add there all your images file extensions

img_type_accepted_by_tf = ["bmp", "gif", "jpeg", "png"]
for filepath in Path(data_dir).rglob("*"):
    if filepath.suffix.lower() in image_extensions:
        img_type = imghdr.what(filepath)
        if img_type is None:
            print(f"{filepath} is not an image")
        elif img_type not in img_type_accepted_by_tf:
            print(f"{filepath} is a {img_type}, not accepted by TensorFlow")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define data augmentation parameters (you can customize these)
train_datagen = ImageDataGenerator(
    rescale=1./255,          # Normalize pixel values to [0, 1]
    rotation_range=20,      # Randomly rotate images by up to 20 degrees
    width_shift_range=0.2,  # Randomly shift images horizontally by up to 20%
    height_shift_range=0.2, # Randomly shift images vertically by up to 20%
    shear_range=0.2,        # Shear transformations
    zoom_range=0.2,         # Randomly zoom in by up to 20%
    horizontal_flip=True,   # Randomly flip images horizontally
    fill_mode='nearest'     # Fill missing pixels with the nearest value
)

# Define the path to your training dataset and the batch size
train_data_dir = 'C:/Users/Deepak Santhosh/Documents/Deep_learning/Plants'
batch_size = 32

# Create the train_generator
train_generator = train_datagen.flow_from_directory(
    train_data_dir,          # Path to the training data directory
    target_size=(256, 256),  # Resize images to the input size expected by MobileNetV2
    batch_size=batch_size,   # Number of images in each batch
    class_mode='categorical' # Specify 'categorical' for multi-class classification
)
```

```python
train_generator.class_indices
```

```python
items=['Neem','Aloe vera','Ashwagandha','Thulasi']
```

```python
num_classes=4
```

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2

# Load MobileNetV2 without top (fully connected) layers
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

# Add Global Average Pooling layer
x = tf.keras.layers.GlobalAveragePooling2D()(base_model.output)

# Add output layer for classification (adjust the number of units)
output = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

# Create the final model
model1 = tf.keras.models.Model(inputs=base_model.input, outputs=output)
```

```python
model1.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model1.fit(train_generator,
                    epochs=5)
```

```python
model1.save('MobileNetV2Model.h5')
```

```python
def predict(image_path):
  img = cv2.imread(image_path)
  resize = tf.image.resize(img,(256,256))
  yhat = model1.predict(np.expand_dims(resize/255,0))
  return np.argmax(yhat),yhat
```
